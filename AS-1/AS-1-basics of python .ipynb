```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
num = int(input("Enter a number: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# Exercise 2:
# Product of Random Numbers
```

```python
import random
def multiplication_quiz():
    num1 = random.randint(1, 450)
    num2 = random.randint(1, 500)
    print(f"What is {num1} × {num2}?")
    user_answer = int(input("Enter your answer: "))
    correct_answer = num1 * num2
    if user_answer == correct_answer:
        print("Correct! Well done.")
    else:
        print(f"incorrect. The correct answer is {correct_answer}.")
multiplication_quiz()
```

```python
# Exercise 3:
# Squares of Even/Odd Numbers
```

```python
def print_even_squares(start=100, end=200):
    for num in range(start, end + 1):
        if num % 2 == 0:  # Check if the number is even
            print(f"{num}^2 = {num ** 2}")
print_even_squares()
```

```python
ROOM_CHARGE_PER_DAY = 2000
DOCTOR_CONSULTATION_FEE = 1500
LAB_TEST_CHARGE = 300

# Patient details (given in the problem)
days_stayed = 4
medicine_charges = 2400
num_tests = 3

# Calculate charges
room_charges = days_stayed * ROOM_CHARGE_PER_DAY
lab_charges = num_tests * LAB_TEST_CHARGE

# Total before discount
total_bill = room_charges + DOCTOR_CONSULTATION_FEE + lab_charges + medicine_charges

# Apply discount if applicable
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount
# Display the bil
print(f"Room Charges ({days_stayed} days): {room_charges}")
print(f"Doctor Consultation Fee: {DOCTOR_CONSULTATION_FEE}")
print(f"Lab Test Charges ({num_tests} tests): {lab_charges}")
print(f"Medicine Charges: {medicine_charges}")
print(f"Total (before discount): {total_bill}")
print(f"Discount Applied: {discount}")
print(f"Final Bill: {final_bill}")
```
